# Przykładowa Analiza Obrazu

Ten notatnik demonstruje podstawowy przepływ pracy z modułem `image_analysis`:
1. Wczytanie i wstępne przetwarzanie obrazu
2. Wykrywanie obiektów
3. Klasyfikacja
4. Wizualizacja wyników

```python
import numpy as np
import cv2
from pathlib import Path

# Import public API
from image_analysis.preprocessing import load_image, resize_image, normalize_image
from image_analysis.detection import detect_objects, draw_bounding_boxes
from image_analysis.classification import classify_image
from image_analysis.utils import setup_logging, get_project_root

setup_logging()
PROJECT_ROOT = get_project_root()
print(f"Katalog projektu: {PROJECT_ROOT}")
```

## 1. Wczytanie i przygotowanie obrazu

Zastąp ścieżkę poniżej ścieżką do swojego pliku.

```python
# Create a synthetic test image when no real image is available
rng = np.random.default_rng(seed=0)
sample_image: np.ndarray = rng.integers(0, 255, (480, 640, 3), dtype=np.uint8)

# Alternatively: load from disk
# sample_image = load_image(PROJECT_ROOT / "data" / "raw" / "sample.jpg")

print(f"Rozmiar obrazu: {sample_image.shape}")
print(f"Typ danych: {sample_image.dtype}")
```

```python
# Resize and normalise
resized = resize_image(sample_image, width=320, height=240)
normalised = normalize_image(resized)

print(f"Po zmianie rozmiaru: {resized.shape}")
print(f"Po normalizacji: dtype={normalised.dtype}, zakres=[{normalised.min():.2f}, {normalised.max():.2f}]")
```

## 2. Wykrywanie obiektów

Aktualnie `detect_objects` zwraca pustą listę (implementacja zastępcza).
Zastąp ją wywołaniem swojego modelu.

```python
detections = detect_objects(resized, confidence_threshold=0.5)
print(f"Liczba wykrytych obiektów: {len(detections)}")
for det in detections:
    print(f"  {det.label}: {det.confidence:.2%} @ {det.bbox}")
```

## 3. Klasyfikacja obrazu

```python
label, confidence = classify_image(resized)
print(f"Przewidywana klasa: '{label}' (pewność: {confidence:.2%})")
```

## 4. Wizualizacja wyników

```python
try:
    import matplotlib.pyplot as plt

    annotated = draw_bounding_boxes(resized, detections)
    # Convert BGR to RGB for matplotlib display
    annotated_rgb = cv2.cvtColor(annotated, cv2.COLOR_BGR2RGB)

    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    axes[0].imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
    axes[0].set_title("Obraz wejściowy")
    axes[0].axis("off")

    axes[1].imshow(annotated_rgb)
    axes[1].set_title(f"Wynik detekcji ({len(detections)} obiektów)")
    axes[1].axis("off")

    plt.tight_layout()
    plt.show()
except ImportError:
    print("Zainstaluj matplotlib: pip install matplotlib")
```
